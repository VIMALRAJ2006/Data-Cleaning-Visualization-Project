```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/Superstore.csv", encoding='latin1')

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df)

plt.title("Sales by Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=df)

plt.title("Profit by Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=df)

plt.title("Sales by Region")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Profit', data=df)

plt.title("Profit by Region")
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_states.plot(kind='bar')

plt.title("Top 10 States by Sales")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30)

plt.title("Sales Distribution")
plt.show()
```

# Key Insights

1. Technology category generated the highest sales.
2. Western region had strong overall performance.
3. Some states contribute significantly more sales than others.
4. Monthly sales show fluctuations across the year.
5. Sales distribution is positively skewed.
6. Profit and sales show moderate correlation.
